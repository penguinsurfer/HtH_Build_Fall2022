```python
import pandas
import matplotlib.pyplot as plt

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

pokemon_data = pandas.read_csv('Pokemon.csv')
pokemon_data
```

```python
legendary_pokemon = pokemon_data[ pokemon_data['Legendary'] == True ]
highest_total = legendary_pokemon.sort_values(by='Total', ascending=False)
highest_total

#I found that both Mega Mewtwos and Mega Rayquaza have the highest base stats in the first six generations
```

```python
#Now I am going to find the average Special attack of all Psychic Pokemon:
import math

avg_psychic_sp_atk_type_one = pokemon_data[pokemon_data['Type 1'] == 'Psychic']['Sp. Atk'].mean() 
avg_psychic_sp_atk_type_two = pokemon_data[pokemon_data['Type 2'] == 'Psychic']['Sp. Atk'].mean()
avg_psychic_sp_atk_type_two + avg_psychic_sp_atk_type_one / 2
```

```python
last_100_pokemon = pokemon_data.tail(100)
last_100_pokemon
```

```python
#Find the fastest speed stats of all Pokemon and list them from fastest to slowest

fastest_pokemon = pokemon_data['Speed'].sort_values(ascending=False)
fastest_pokemon
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression

x = pokemon_data['Attack'].values.reshape(-1, 1)
y = pokemon_data['Defense']

lr_model = LinearRegression()
lr_model.fit(x, y)
y_pred = lr_model.predict(x)

plt.scatter(x, y)
plt.xlabel('Attack')
plt.ylabel('Defense')

plt.plot(x, y_pred)
```
